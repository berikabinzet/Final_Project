```python
#importing data
data = pd.read_csv("src/original_data/child_ineq_data.csv")
cid_var_info = pd.read_stata("src/original_data/var_info.dta")
```

```python
#renaming variables and dropping columns to match
cid_var_info = cid_var_info.astype({'survey_year':'str'})
cid_var_info["var_name"]=cid_var_info["readable_name"]+cid_var_info["survey_year"]
data = data.filter(items=cid_var_info["nlsy_name"], axis=1)
data.rename(columns = dict(zip(cid_var_info["nlsy_name"],cid_var_info["var_name"])), inplace = True)
```

```python
#replacing negative values 
data = data.replace(list(range(-10,0)),np.nan)
```
